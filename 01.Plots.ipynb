```python
#wave speed
v=0.01
#wave length
l=2*np.pi
#Wave number:
k=2*np.pi/l
#angular frequency:
w=v*k

# Define the function of time
def graficar(t):
    y = np.sin(k*x-w*t)
    linea.set_data((x,y))
    return (linea, )
```

```python
# x coordinate
x = np.linspace(0, 2*np.pi, 100)

#Define plot, labels, and axes limits
fig, ax = plt.subplots()
ax.set_xlabel('X')
ax.set_ylabel('$D(x,t)$')
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.1, 1.1)

#Define the line properties
linea, = ax.plot([],[],'-',color = 'b', label = 'prueba')

#Make the animation
anim = animation.FuncAnimation(fig, graficar, frames=300, interval=10)
anim
```
